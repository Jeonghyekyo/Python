```python
#2번

import numpy as np
import matplotlib.pyplot as plt
#import math
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

def f(a,b):
	return a**2 + b**2 +2* np.sin(a*b)

def grad_f(a,b): 
  h=1e-5
  return [(f(a+h,b) - f(a-h,b))/(2*h),(f(a,b+h) - f(a,b-h))/(2*h)]


def gradient_descent(theta0, iters, alpha):
	history = [theta0] 
	theta = theta0    
	# main loop by iterations:
	for i in range(iters):
		gradient = grad_f(theta[0],theta[1]) 
		# update parameters:
		theta = [a - alpha*b for a,b in zip(theta, gradient)]
		history.append(theta)
	return history

history = gradient_descent(theta0 = [-3, -3], iters =30, alpha = 0.04)


fig = plt.figure(figsize=(20, 15))
ax = fig.gca(projection='3d')

a = np.arange(-5, 5, 0.25)
b = np.arange(-5, 5, 0.25)
a, b = np.meshgrid(a, b)
c = f(a,b)
surf = ax.plot_surface(a, b, c, rstride=1, cstride=1, alpha=0.1, 
		 linewidth=0, antialiased=False, cmap='rainbow')


a = np.array([x[0] for x in history])
b = np.array([x[1] for x in history])
c =  f(a,b)

ax.scatter(a, b, c, color="r"); 

print(c)


plt.xlabel('Feature A')
plt.ylabel('Feature B')

plt.show()
```
