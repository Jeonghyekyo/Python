```python
#202203280 정혜교

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import pandas as pd
import matplotlib.pyplot as plt

targetUrl_Caffeine="/content/caffeine.csv"

df = pd.read_csv(targetUrl_Caffeine)

normalization_df = (df - df.mean())/df.std()
X=normalization_df.iloc[:,0:3]

y = df.iloc[:,4]

y = pd.get_dummies(y)

model = Sequential()
model.add(Dense(60,  input_dim=3, activation='relu'))
model.add(Dense(70,  activation='relu'))
model.add(Dense(6, activation='softmax'))

model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])

history = model.fit(X, y, epochs=10000, batch_size=10)

loss = model.evaluate(X,y)

print(loss)
```

```python
import numpy as np

y_loss = history.history['loss']

x_len = np.arange(len(y_loss))

plt.plot(x_len,y_loss)

y_accuracy = history.history['accuracy']

x_len = np.arange(len(y_accuracy))

plt.plot(x_len,y_accuracy)
```
